```python
import os
import requests
import datetime
```

```python
# your AgroDataCube access token - replace this code or add the OS environment variable
token = os.environ["AGRODATACUBE_TOKEN"]

request_header = {"token":token}
```

```python
cropcodes_url = "http://agrodatacube.wur.nl/api/v1/rest/codes/cropcodes?page_size=1000&page_offset=0"
cropcodes_data = requests.get(cropcodes_url, headers=request_header)
cropcodes = cropcodes_data.json()
len(cropcodes["features"])
```

```python
for cropcode in cropcodes["features"]:
    code = cropcode["properties"]["cropcode"]
    name = cropcode["properties"]["cropname"]
    crop_id = cropcode["properties"]["cropid"]
    print(code, name)
```
